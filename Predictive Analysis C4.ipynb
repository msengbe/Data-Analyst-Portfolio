Select relevant columns and drop any rows with missing values

```python
selected_features = [
    'Tenure', 'Bandwidth_GB_Year', 'Age', 'Income', 'Children',
    'Outage_sec_perweek', 'Yearly_equip_failure',
    'Contract', 'InternetService', 'PaymentMethod',
    'Gender', 'Marital', 'Techie', 'DeviceProtection',
    'OnlineSecurity', 'OnlineBackup', 'StreamingTV',
    'MonthlyCharge'
]
```

```python
df_selected = df[selected_features].dropna()
```

Separate predictors (X_raw) and target (y)

```python
X_raw = df_selected.drop('MonthlyCharge', axis=1)
y = df_selected['MonthlyCharge'].copy()
```

One-hot encode categorical variables 

```python
categorical_features = [
    'Contract', 'InternetService', 'PaymentMethod',
    'Gender', 'Marital', 'Techie', 'DeviceProtection',
    'OnlineSecurity', 'OnlineBackup', 'StreamingTV'
]
```

```python
X_encoded = pd.get_dummies(X_raw, columns=categorical_features, drop_first=True)
```

Standardize numeric features to zero mean and unit variance

```python
numeric_features = [
    'Tenure', 'Bandwidth_GB_Year', 'Age', 'Income',
    'Children', 'Outage_sec_perweek', 'Yearly_equip_failure'
]
scaler = StandardScaler()
```

```python
X_encoded[numeric_features] = scaler.fit_transform(X_encoded[numeric_features])
```

Combine the processed predictors with the target into a final cleaned DataFrame

```python
df_cleaned = pd.concat([X_encoded, y.reset_index(drop=True)], axis=1)
```
